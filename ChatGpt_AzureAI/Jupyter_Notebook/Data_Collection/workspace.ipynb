```python
# Initialize counter
count = 0
# Loop through columns in the input DataFrame
for col1 in input_df.columns:
    # Loop through columns in the mapping DataFrame
    for col2 in mapping_df['Source']:
        # Check if the current column in the input DataFrame matches any Source column value in the mapping DataFrame
        if col1 == col2:
            # If there is a match, increment the counter and print a match message
            count += 1
            print("Matched ::: ", col1, " >> ", col2)
        else:
            # If no match, continue to the next iteration
            pass
count
```

```python
# Initialize an empty dictionary to store the mapping of target columns to corresponding source columns
dict1 = {}
# Loop through rows in the input DataFrame (input_df)
for i in range(input_df.shape[0]):    
    # Loop through columns in the input DataFrame
    for col1 in input_df.columns:        
        # Loop through columns in the 'Source' column of the mapping DataFrame using enumerate to also get the index (ind2)
        for ind2, col2 in enumerate(mapping_df['Source']):            
            # Check if the current column in the input_df matches any column in the mapping DataFrame
            if col1 == col2:
                try:
                    # If a match is found, update the dictionary with the target column as the key
                    # and the corresponding source column's data for the current row
                    dict1.update({mapping_df['Target'][ind2]: input_df[mapping_df['Source'][ind2]]})                    
                except Exception as e:
                    # If an exception occurs (e.g., an issue with updating the dictionary),
                    # print the exception for debugging
                    print(e)
# The dictionary (dict1) now contains the mapping of target columns to corresponding source columns for each row
# You can use dict1 for further processing or analysis                    
```
