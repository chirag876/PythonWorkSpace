we have headers as column names, alter_names as their alternative name.  datas is their respective data.

```python

pairs = []
for i in range(len(list_alt)):
    data_data = ''
    for j in range(len(datas)):
        data_data = data_data+','+datas[j][i]
        #data_data = data_data.replace(data_data[0], '')
    prompt = (
            f"""You will be given a Dataframe with input column and data associated with the columns.\n"""
            f"""Examine the dataset with column name '{headers[i]}'. This would be your target column.\n """
            f"""Alternative names for this column include: '{list_alt[i]}'.\n"""
            f"""Consider the data provided '{data_data}' present in the '{headers[i]}' target column to learn the patterns and characteristics.\n"""
            f"""Attempt to discern any discernible patterns, such as data types, common values, or trends.\n"""
            f"""Pay attention to any potential relationships between data points within this column.\n"""
            f"""If there are null values, feel free to disregard them in your analysis.\n"""
            f"""Your goal is to find the relationship between the input column and target column based on either the data or the column name and return the closest match based on your analysis."""
            f"""If input column and target column are same, then the 'confidence score' should be evaluated as 100. If a relationship between input column and target column cannot be established, then the 'confidence score' is 0. Return the confidence score"""
        )
    
        
    completion = (
            f"""Based on the analysis, the provided entry is likely associated with the '{headers[i]}' column. Confidence score is: 100 \n"""  
            f"""Based on the analysis, the provided entry is not likely associated with the '{headers[i]}' column. Confidence score is: 0"""
    )
    pairs.append({"prompt":prompt, "completion":completion})
```
