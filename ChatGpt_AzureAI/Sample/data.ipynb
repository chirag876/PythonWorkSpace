```python
import pandas as pd
import openpyxl
```

```python
workbook_url ='InspectionResponse_VYRD FNOL Inspection_638257809345214699 1.xls'
```

```python
single_df =  pd.read_excel(workbook_url,sheet_name=None, dtype=str,engine='openpyxl')
for sheet_name, df in single_df.items():
    print(f"Sheet name: {sheet_name}")
    print(df)  # This will print the contents of the dataframe for the current sheet
    print("=" * 40)
```

```python
df_list = []

# Loop through each sheet and add its dataframe to the list
for df in single_df.values():
    df_list.append(df)

# Concatenate dataframes along columns (horizontally)
merged_df = pd.concat(df_list, axis=1)

# Print the merged dataframe
print(merged_df)

# Save the merged dataframe to a new Excel file
merged_df.to_excel('horizontally_merged_output.xlsx', index=False)
```
